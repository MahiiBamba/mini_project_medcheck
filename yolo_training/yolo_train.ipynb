```python
from ultralytics import YOLO
import cv2

# Load trained YOLO model
model = YOLO("runs/detect/train/weights/best.pt")

# Input video
video_path = "/home/mansi/Desktop/mini_project/real-time/test_4.mp4"

# Open video
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Output video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("output_blister_detection-4.mp4", fourcc, fps, (frame_width, frame_height))

while cap.isOpened():
    
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection
    results = model(frame, conf=0.4)

    # Draw bounding boxes
    annotated_frame = results[0].plot()

    # Show video
    cv2.imshow("Blister Cavity Detection", annotated_frame)

    # Save video
    out.write(annotated_frame)

    # Press q to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release everything
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load trained YOLO model
model = YOLO("runs/detect/train/weights/best.pt")

# Open laptop webcam
cap = cv2.VideoCapture(0)

# Set resolution (optional but helps speed)
cap.set(3, 640)
cap.set(4, 480)

while True:
    
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection
    results = model(frame, conf=0.4)

    # Draw bounding boxes
    annotated_frame = results[0].plot()

    # Show live detection
    cv2.imshow("Real-Time Blister Detection", annotated_frame)

    # Press q to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
